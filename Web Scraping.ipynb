```python
from bs4 import BeautifulSoup as bs4
import requests
import pandas as pd

pages=[]
prices=[]
stars=[]
titles=[]
urlss=[]

# I used this website for this project: http://books.toscrape.com 
```

```python
# This code is just to increment the page

pages_to_scrape=5

for i in range(1,pages_to_scrape+1):
    url = ('http://books.toscrape.com/catalogue/page-{}.html').format(i)
    pages.append(url)
#print(pages)

# This code is just parsing text

for item in pages:
    page = requests.get(item)
    soup = bs4(page.text, 'html.parser')
    #print(soup.prettify())
        
## This code is to get the book titles 
    
    for i in soup.findAll('h3'):
        ttl=i.getText()
        #print(ttl)
        titles.append(ttl)
    
## This code is to get the book price
    
    for j in soup.findAll('p', class_='price_color'):
        price=j.getText()
        prices.append(price)

## This code is to get the star rating

    for s in soup.findAll('p', class_='star-rating'):
        for k,v in s.attrs.items():
            star =v[1]
            stars.append(star) 

## This code is to get the book image URL 

    divs =soup.findAll('div', class_='image_container')
    for thumbs in divs:
        tgs=thumbs.find('img',class_='thumbnail')
        urls='http://books.toscrape.com/'+str(tgs['src'])
        newurls=urls.replace("../","")
        urlss.append(newurls)
        
```

```python
data={'Title': titles, 'Prices': prices, 'Stars':stars, "URLs":urlss}
#data
```

```python
df=pd.DataFrame(data=data)
df.index+=1
```
